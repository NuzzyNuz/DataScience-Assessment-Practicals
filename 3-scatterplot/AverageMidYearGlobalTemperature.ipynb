```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
csvData = pd.read_csv('GISSSurfaceTemperatureAnalysis.csv')
csvData
```

```python
csvData.describe()
```

```python
y = csvData['Jun']
x1 = csvData['Year']
```

```python
plt.scatter(x1,y)
plt.title("Annual Mid Year Global Temperature", size=20)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Temperature (celcius)',fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0072*x1 + -13.9963
fig = plt.plot(x1,yhat, lw=4, c='orange',label='regression line')
plt.title("Annual Mid Year Global Temperature", size=20)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Temperature (celcius)',fontsize=20)
plt.show()
```
